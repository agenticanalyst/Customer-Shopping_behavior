```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
```

```python
# 1. Purchase Month
df['Purchase Month'] = df['Purchase Date'].dt.month_name()
```

```python
# 2. Purchase Year
df['Purchase Year'] = df['Purchase Date'].dt.year
```

```python
# 3. Purchase Day
df['Purchase Day'] = df['Purchase Date'].dt.day_name()
```

```python
# 4. Age Group
df['Age Group'] = pd.cut(
    df['Age'],
    bins=[0,18,25,35,45,60,100],
    labels=['<18','18-25','26-35','36-45','46-60','60+']
)
```

```python
# 5. Spending Category
df['Spending Category'] = pd.cut(
    df['Purchase Amount (â‚¹)'],
    bins=[0,1000,3000,6000,100000],
    labels=['Low','Medium','High','Premium']
)
```

```python
# 6. Total Cost (Approx)
df['Total Amount'] = (
    df['Purchase Amount (â‚¹)'] * df['Quantity']
) - (
    df['Purchase Amount (â‚¹)'] * df['Quantity'] * df['Discount (%)'] / 100
)
```

```python
# 7. Delivery Category
df['Delivery Category'] = pd.cut(
    df['Delivery Time (Days)'],
    bins=[0,2,5,10],
    labels=['Fast','Normal','Slow']
)
```

```python
# 2. Average Purchase by Gender
gender_avg = df.groupby('Gender')['Purchase Amount (Rs)'].mean()

print(gender_avg)

plt.figure(figsize=(6,4))
gender_avg.plot(kind='bar')
plt.title("Average Purchase by Gender")
plt.ylabel("Average Purchase")
plt.show()
```

```python
# 3. Sales by Payment Method

payment_sales = df.groupby('Payment Method')['Purchase Amount (Rs)'].sum()

print(payment_sales)

plt.figure(figsize=(8,5))
payment_sales.plot(kind='bar')
plt.title("Sales by Payment Method")
plt.ylabel("Total Sales")
plt.xticks(rotation=30)
plt.show()
```

```python
# 4. Sales by Location
location_sales = df.groupby('Location')['Purchase Amount (Rs)'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))
location_sales.plot(kind='bar')
plt.title("Sales by Location")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Average Rating by Category
rating = df.groupby('Category')['Review Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
rating.plot(kind='bar')
plt.title("Average Review Rating by Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```
